```python
import numpy as np
#output_file("bar_colormapped.html")

from bokeh.core.properties import value

from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import ColumnDataSource

from bokeh.transform import dodge



avg=[  webs['(Fully loaded) Requests'].mean(),webs['DOM elements'].mean(), webs['First byte'].mean(), webs['Load time'].mean(), webs['Speed Index'].mean(), webs['Start render'].mean()]
par_values =[input_web[4], input_web[6],input_web[7],input_web[8],input_web[9],input_web[10]]
# par_diff = [y-x for y,x  in zip(avg, par_value)]
params = [ 'Requests', 'DOM elements', 'First byte', 'Load time', 'Speed Index', 'Start render']
values = ["Standard", "input"]
colors = ["#c9d9d3", "#718dbf"]




data = {'params' : params,
        'Standard'   : avg,
        'input'   : par_values,
        }
source = ColumnDataSource(data=data)
# p.vbar_stack(years, x='fruits', width=0.9, color=colors, source=data, legend=[value(x) for x in years])

# p.y_range.start = 0
# p.x_range.range_padding = 0.1
# p.xgrid.grid_line_color = None
# p.axis.minor_tick_line_color = None
# p.outline_line_color = None
# p.legend.location = "top_left"
# p.legend.orientation = "horizontal"

# p.vbar(x=dodge('fruits', -0.25, range=p.x_range), top='2015', width=0.2, source=source,color="#c9d9d3", legend=value("2015"))

p = figure(x_range=params, plot_height=450, toolbar_location=None, title="Average Performance of Website parameters")
# p.vbar_stack(values, x="params", width=0.9,source=data, legend=[value(x) for x in values], color=colors, alpha=0.5) #, line_color='white')
p.vbar(x=dodge('params',-0.15,range=p.x_range), top = 'Standard', width=0.25, source=source, color = "#c9d9d3", legend=value("Standard"))
p.vbar(x=dodge('params',0.15,range=p.x_range), top = 'input', width=0.25, source=source, color = "#718dbf", legend=value("input cite"))
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.y_range.end = 1
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

show(p)
```
